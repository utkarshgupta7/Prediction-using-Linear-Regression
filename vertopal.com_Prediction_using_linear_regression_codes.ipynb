    def normalize(df1):
        result = df1.copy()
        for feature_name in df1.columns:
            max_value = df1[feature_name].max()
            min_value = df1[feature_name].min()
            result[feature_name] = (df1[feature_name] - min_value) / (max_value - min_value)
        return result
